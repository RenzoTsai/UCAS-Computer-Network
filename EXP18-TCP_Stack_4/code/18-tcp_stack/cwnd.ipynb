```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd
```

```python
def handle_cwnd(path):
    data = pd.read_csv(path)
    ydata = []
    xdata = []
    xdata = data.ix[:,'time']
    ydata = data.ix[:,'cwnd']
    return xdata, ydata
```

```python
def print_plt(x, y, type):
    plt.plot(x, y, color = 'green')
    plt.margins(0)
    plt.xlabel("time")
    plt.ylabel(type)
    plt.title(type)
    plt.tick_params(axis="both")
    plt.legend()
    plt.show()
```

```python
    input_path = "cwnd.csv"
    x, y = handle_cwnd(input_path)
    print_plt(x,y, "CWND")
```
